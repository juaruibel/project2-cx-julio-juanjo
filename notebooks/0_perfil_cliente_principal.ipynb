### Definimos características del cliente principal

#### Justificación del filtro `high_value_client`

- En el notebook definimos un umbral con `q1 = df_client_mode["bal"].quantile(0.25)`. Es el **percentil 25** (Q1).
- A partir de ese umbral, filtramos con `high_value_client = df_client_mode[df_client_mode["bal"] > q1]`.

¿Qué significa esto exactamente?
- Nos estamos quedando con los clientes cuyo `bal` está **por encima del percentil 25**.
- Es decir: **excluimos el 25% con menor balance** y nos quedamos con el **75% restante**.

Por qué lo hacemos así:
- No buscamos solo “los más ricos” (top 25%), sino evitar que el análisis se vea dominado por clientes con balance muy bajo (que pueden distorsionar la segmentación y los gráficos).
- Con este criterio nos quedamos con un grupo “de valor razonable” (por encima de P25), suficientemente amplio para que el EDA sea estable y a la vez eliminando el cuartil inferior.

Nota:
- Si quisiéramos un “high value” más estricto (top 25%), entonces el umbral correcto sería `quantile(0.75)` y el filtro sería `bal > P75`.

## Resumen/Conclusiones

### Resumen

- Identificamos el perfil del cliente principal utilizando la **moda** de `logons_6_mnth`, ya que representa el comportamiento más frecuente de uso del proceso online. En nuestro caso, la moda coincide con el **valor máximo** observado en la variable (`9`) y filtramos clientes con `logons_6_mnth >= moda`.
- Aplicamos el mismo criterio a `calls_6_mnth`, donde la moda también coincide con el **valor máximo** (`6`) y filtramos clientes con `calls_6_mnth >= moda`. Con ello, el “cliente principal” queda caracterizado como el usuario más común en actividad digital y contacto con el servicio.
- Para contextualizar la relevancia económica del perfil, incorporamos `bal` (balance) y filtramos por encima del **percentil 25** (`bal > P25`), es decir, trabajamos con el **75% superior** de balances, excluyendo el cuartil inferior.
- Finalmente, para estructurar el análisis demográfico y facilitar la interpretación, agrupamos a los clientes en categorías discretas: `Age_category` (rangos de edad) y `Years_category` (tipología por antigüedad en la plataforma).

### Conclusiones (a partir de los insights)

- El perfil del cliente principal se concentra en edades altas: `Old` domina claramente frente a `Medium` y `Young`.
- Por tenencia, predomina `New` (2–15 años), seguido de `Established` (15–35 años), mientras que `Loyal` es residual.
- En el cruce `Age_category` x `Years_category`, `Old` sigue siendo la categoría dominante incluso dentro de `New`, lo que sugiere que el cliente típico no es necesariamente joven aunque su cuenta sea relativamente reciente.
